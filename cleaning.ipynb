```python
for i in mental_health.columns:
    if mental_health[i].dtype == 'object':
        print(f'{i}的唯一值：{mental_health[i].unique()}')
```

```python
data = mental_health.loc[:,'age':]
data.head()
```

```python
# 将object转为int
for col in data:
    if data[col].dtype == 'object':
        uni = data[col].unique()
        def convert(item):
            index = np.argwhere(uni == item)[0][0]
            return index
        data[col] = data[col].map(convert)
```

```python
data
```

```python
data.isnull().any()
```

```python
#  假设有缺失值
# 处理方法一 删除缺失值
# 修改原表使用inplace=True
data1 = data.dropna(subset=['age'])
data1
```

```python
# 处理方二填充缺失值

data2 = data.fillna({'age': data['age'].mean()})
data2
```
