```python
import requests # 发送http请求 调用天气api
import argparse # 解析命令行参数
import os # 获取环境变量
```

```python
# API_KEY = os.getenv('API_KEY')
API_KEY = os.getenv("API_KEY")
API_KEY
# if not API_KEY:
#     raise ValueError("请设置环境变量API_KEY，获取天气API的访问权限")
```

```python
# 定义实时天气接口(固定的 API Host)
location = "101010100"
BASE_URL = f"https://jq3fc2277w.re.qweatherapi.com/v7/weather/now?location={location}&key={API_KEY}"
BASE_URL
```

```python
# 定义获取天气信息函数
def get_weather(city):
    location_url = BASE_URL.format(location=city, API_KEY=API_KEY)
    location_params = {
        'location': city,
        'key': API_KEY,
        'lang': 'zh',
        'unit': 'm'
    }
    # 异常处理
    try:
        response = requests.get(location_url, params=location_params)
        location_data = response.json()
        if location_data['code'] == '200':
            city_id = location_data['location'][0]['id']
            city_name = location_data['location'][0]['name']
            weather_params = {
                'location': city_id,
                'key': API_KEY,
                'lang': 'zh',
                'unit': 'm'
            }
            weather_response = requests.get(BASE_URL,params=weather_params)
            weather_data = weather_response.json()
            if weather_data['code'] == '200':
                now = weather_data['now']
                return {
                    'city':city_name,
                    'temp':now['temp'],
                    'text':now['text'],
                    'wind_dir':now['windDir'],
                    'humidity':now['humidity']
                }
            else:
                return f"获取天气失败:{weather_data['code']}"
        else:
            return f'未找到城市:{city}'
    except Exception as e :
        return f'请求失败:{str(e)}'
```

```python
# 24. 定义格式化输出函数，让天气信息显示更美观
def format_output(weather_info):
    """格式化输出天气信息"""  # 25. 函数文档字符串
    # 26. 判断传入的参数是否是字典（即请求成功的情况）
    if isinstance(weather_info, dict):
        # 27. 拼接成美观的字符串（用emoji符号提升可读性）
        return (
            f"📍 {weather_info['city']}\n"          # 城市名 + 定位图标
            f"🌡️  温度: {weather_info['temp']}°C\n" # 温度 + 温度计图标
            f"☁️  天气: {weather_info['text']}\n"   # 天气状况 + 云朵图标
            f"💨 风向: {weather_info['wind_dir']} {weather_info['wind_scale']}级\n" # 风向风力 + 风图标
            f"💧 湿度: {weather_info['humidity']}%" # 湿度 + 水滴图标
        )
    else:
        # 28. 如果是字符串（即请求失败的情况），直接返回错误信息
        return weather_info
```

```python
def main():
    # 30. 创建参数解析器对象，描述工具的作用
    parser = argparse.ArgumentParser(description="CLI 天气查询工具")
    # 31. 添加必选参数city，说明该参数是要查询天气的城市名称
    # parser.add_argument("city", help="要查询天气的城市名称")
    parser.add_argument("city", nargs='?', default="北京", help="要查询天气的城市名称")
    # 32. 解析用户在终端输入的参数（比如输入`python weather.py 北京`，就会解析出city=北京）
    args, unknown = parser.parse_known_args()
    
    # 33. 调用get_weather函数，传入解析到的城市名，获取天气信息
    weather_info = get_weather(args.city)
    # 34. 调用format_output函数格式化数据，并打印到终端
    print(format_output(weather_info))

if __name__ == "__main__":
    # 36. 执行主函数，启动程序
    main()
```
