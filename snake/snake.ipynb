```python
import pygame
import random

pygame.init()
W = 800
H = 600
ROW = 60
COL = 80
size = (W, H)


class Point:
    row = 0
    col = 0

    def __init__(self, row, col):
        self.row = row
        self.col = col

    def copy(self):
        return Point(row=self.row, col=self.col)


window = pygame.display.set_mode(size)
pygame.display.set_caption('贪吃蛇')
snake = []

def food_spawn():
    while True:
        cond = False
        food_pos = Point(row=random.randint(0, ROW - 1), col=random.randint(0, COL - 1))
        if head.row == food_pos.row and head.col == food_pos.col:
            cond = True
        for i in snake:
            if i.row == food_pos.row and i.col == food_pos.col:
                cond = True
                break
        if not cond:
            break
    return food_pos


# 初始坐标
head = Point(row=int(ROW / 2), col=int(COL / 2))
head_color = (0, 47, 167)
food = food_spawn()
food_color = (255, 230, 0)
direct = 'left'



def rect(point, color):
    cell_wid = W / COL
    cell_ht = H / ROW
    left = point.col * cell_wid
    top = point.row * cell_ht
    pygame.draw.rect(window, color, (left, top, cell_wid, cell_ht))


quit = True
clock = pygame.time.Clock()
while quit:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            quit = False

        elif event.type == pygame.KEYDOWN:
            print(event)
            if event.key == 119:
                if direct == 'left' or direct == 'right':
                    direct = 'up'
            elif event.key == 115:
                if direct == 'left' or direct == 'right':
                    direct = 'down'
            elif event.key == 97:
                if direct == 'up' or direct == 'down':
                    direct = 'left'
            elif event.key == 100:
                if direct == 'up' or direct == 'down':
                    direct = 'right'
    eat = (head.row == food.row and head.col == food.col)
    if eat:
        food = food_spawn()
        snake.insert(0, head.copy())
    else:
        snake.insert(0, head.copy())
        snake.pop()
    if direct == 'left':
        head.col -= 1
    elif direct == 'right':
        head.col += 1
    elif direct == 'up':
        head.row -= 1
    elif direct == 'down':
        head.row += 1
    dead = False
    if head.row < 0 or head.col < 0 or head.row >= ROW or head.col >= COL:
        dead = True
    for i in snake:
        if head.col == i.col and head.row == i.row:
            dead = True
            break
    if dead:
        print('游戏结束')
        quit = False
    pygame.draw.rect(window, (255, 255, 255), (0, 0, W, H))
    rect(head, head_color)
    rect(food, food_color)
    for i in snake:
        rect(i, head_color)

    # 交还控制权给系统
    pygame.display.flip()
    clock.tick(90)
```
